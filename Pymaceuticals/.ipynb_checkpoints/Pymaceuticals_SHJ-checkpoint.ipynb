```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
drug_data = "data/mouse_drug_data.csv"
trial_data = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data

drugfile = pd.read_csv(drug_data)
drugfile.head()
```

```python
drugfile.count()
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data

trialfile = pd.read_csv(trial_data)
trialfile.head()
```

```python
trialfile.count()
```

```python
# Combine the data into a single dataset
# Display the data table for preview

combinedfile = pd.merge(trialfile, drugfile, on ="Mouse ID", how="left")
combinedfile.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

meantumor = combinedfile.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
# Preview DataFrame

meantumordf = meantumor.reset_index()
meantumordf = meantumordf.rename({"Tumor Volume (mm3)" : "Tumor Volume (mm3) MEAN"}, axis='columns')
meantumordf.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

semtumor = combinedfile.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()
semtumor.head()
```

```python
# Convert to DataFrame
# Preview DataFrame

semtumor = pd.DataFrame(semtumor)
semtumor = semtumor.rename({"Tumor Volume (mm3)" : "Tumor Volume (mm3) SEM"}, axis='columns')
semtumordf = semtumor.reset_index()
semtumordf.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

newmeantumor = meantumordf.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3) MEAN")
newmeantumor
```

```python
newmeantumor.index
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure

capomulin = semtumordf.loc[semtumordf["Drug"]=="Capomulin"]
infubinol = semtumordf.loc[semtumordf["Drug"]=="Infubinol"]
ketapril = semtumordf.loc[semtumordf["Drug"]=="Ketapril"]
placebo = semtumordf.loc[semtumordf["Drug"]=="Placebo"]

plt.errorbar(newmeantumor.index, newmeantumor["Capomulin"], capomulin["Tumor Volume (mm3) SEM"],
            fmt="o", color="red", ls="--", linewidth=0.5, alpha=0.9, label="Capomulin")
plt.errorbar(newmeantumor.index, newmeantumor["Infubinol"], infubinol["Tumor Volume (mm3) SEM"],
            fmt="^", color="blue", ls="--", linewidth=0.5, alpha=0.9, label="Infubinol")
plt.errorbar(newmeantumor.index, newmeantumor["Ketapril"], ketapril["Tumor Volume (mm3) SEM"],
            fmt="s", color="green", ls="--", linewidth=0.5, alpha=0.9, label="Ketapril")
plt.errorbar(newmeantumor.index, newmeantumor["Placebo"], placebo["Tumor Volume (mm3) SEM"],
            fmt="d", color="black", ls="--", linewidth=0.5, alpha=0.9, label="Placebo")

plt.legend(loc="best")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()


plt.savefig("response.png")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint  

metastatic = combinedfile.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
# Preview DataFrame
metastaticdf = pd.DataFrame(metastatic)
metastaticdf = metastaticdf.reset_index()
metastaticdf.head(10)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

metastaticsem = combinedfile.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
# Preview DataFrame
metastaticsemdf = pd.DataFrame(metastaticsem)
metastaticsemdf = metastaticsemdf.reset_index()
metastaticsemdf.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

newmetastaticdf = metastaticdf.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")
newmetastaticdf
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure

capomulin = metastaticsemdf.loc[metastaticsemdf["Drug"]=="Capomulin"]
infubinol = metastaticsemdf.loc[metastaticsemdf["Drug"]=="Infubinol"]
ketapril = metastaticsemdf.loc[metastaticsemdf["Drug"]=="Ketapril"]
placebo = metastaticsemdf.loc[metastaticsemdf["Drug"]=="Placebo"]

plt.errorbar(newmetastaticdf.index, newmetastaticdf["Capomulin"], capomulin["Metastatic Sites"],
            fmt="o", color="red", ls="--", linewidth=0.5, alpha=0.9, label="Capomulin")
plt.errorbar(newmetastaticdf.index, newmetastaticdf["Infubinol"], infubinol["Metastatic Sites"],
            fmt="^", color="blue", ls="--", linewidth=0.5, alpha=0.9, label="Infubinol")
plt.errorbar(newmetastaticdf.index, newmetastaticdf["Ketapril"], ketapril["Metastatic Sites"],
            fmt="s", color="green", ls="--", linewidth=0.5, alpha=0.9, label="Ketapril")
plt.errorbar(newmetastaticdf.index, newmetastaticdf["Placebo"], placebo["Metastatic Sites"],
            fmt="d", color="black", ls="--", linewidth=0.5, alpha=0.9, label="Placebo")

plt.legend(loc="best")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Durationn (Days)")
plt.ylabel("Met. Sites")
plt.grid()


plt.savefig("metastatic.png")
plt.show()

```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mousecount = combinedfile.groupby(["Drug", "Timepoint"])["Mouse ID"].nunique()

# Convert to DataFrame
# Preview DataFrame
mousecountdf = pd.DataFrame(mousecount)
mousecountdf = mousecountdf.reset_index()
mousecountdf.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame

newmousecountdf = mousecountdf.pivot(index="Timepoint", columns="Drug", values="Mouse ID")
newmousecountdf.head(10)
```

```python
# Generate the Plot (Accounting for percentages)
# Save the Figure
# Show the Figure

capomulin = newmousecountdf["Capomulin"]/25*100
infubinol = newmousecountdf["Infubinol"]/25*100
ketapril = newmousecountdf["Ketapril"]/25*100
placebo = newmousecountdf["Placebo"]/25*100

plt.plot(newmousecountdf.index, capomulin, marker="o", color="red", ls="--", linewidth=0.5, alpha=0.9, label="Capomulin")
plt.plot(newmousecountdf.index, infubinol, marker="^", color="blue", ls="--", linewidth=0.5, alpha=0.9, label="Infubinol")
plt.plot(newmousecountdf.index, ketapril, marker="s", color="green", ls="--", linewidth=0.5, alpha=0.9, label="Ketapril")
plt.plot(newmousecountdf.index, placebo, marker="d", color="black", ls="--", linewidth=0.5, alpha=0.9, label="Placebo")

plt.legend(loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Times (Days)")
plt.ylabel("Survital Rate(%)")
plt.grid()

plt.savefig("survival.png")
plt.show()

```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
pctchange = (newmeantumor.iloc[-1]/newmeantumor.iloc[0]-1)*100
pctchange
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
# Orient widths. Add labels, tick marks, etc. 
drug = ("Capomulin","Infubinol","Ketapril","Placebo")
pctchange = (-19.48, 46.12, 57.03, 51.30)

passing=["Capomulin"]
faling=["Infubinol","Ketapril","Placebo"]

# Use functions to label the percentages of changes
# Call functions to implement the function calls

summarybar = plt.bar(drug, pctchange, color=["green", "red", "red", "red"], 
                     width=1, alpha=0.9, align="edge")

tick_locations=[value+1 for value in np.arange(len(drug))]
plt.xticks(tick_locations, drug)

plt.grid()
plt.ylim(-30,70)
plt.ylabel("% Tumor Volume Change")
plt.xlabel("Drug")
plt.title("Tumor Change over 45 Day Treatment")


# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            label_position = 5
        else:
            label_position = -10
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')
                
# Call functions to implement the function calls
autolabel(summarybar)

# Save the Figure
# Show the Figure

plt.savefig("tumorsummary.png")
plt.show()

```
